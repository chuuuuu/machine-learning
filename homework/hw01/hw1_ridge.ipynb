```python
import csv
import numpy as np

TRAIN_PATH: f'{DATA_PATH}/covid.train.csv'
TEST_PATH: f'{DATA_PATH}/covid.test.csv'

with open(TRAIN_PATH, 'r') as f:
    train_data = list(csv.reader(f))
    train_data = np.array(train_data[1:])[:, 1:].astype(np.float32)

train_X = train_data[:, :-1]
train_y = train_data[:, -1]

with open(TEST_PATH, 'r') as f:
    test_data = list(csv.reader(f))
    test_data = np.array(test_data[1:])[:, 1:].astype(np.float32)

test_X = test_data
```

```python
from sklearn.feature_selection import f_regression, SelectKBest, mutual_info_regression

INPUT_DIM = 14

selector = SelectKBest(f_regression, k=INPUT_DIM)
selector.fit(train_X, train_y)
cols = selector.get_support(indices=True)
print(cols)

train_X = train_X[:, cols]
test_X = test_X[:, cols]
```

```python
# from sklearn.linear_model import LinearRegression
# reg = LinearRegression().fit(train_X, train_y)
# reg.score(train_X, train_y)
# test_y = reg.predict(test_X)

from sklearn.linear_model import Ridge
clf = Ridge(alpha=1, tol=1e-8)
clf.fit(train_X, train_y)
print(clf.coef_)
test_y = clf.predict(test_X)

# from sklearn.linear_model import RidgeCV
# clf = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1, 10]).fit(train_X, train_y)
# test_y = clf.predict(test_X)
# print(f'clf.alpha_: {clf.alpha_}')
```

```python
PRED_PATH = 'pred.csv'

with open(PRED_PATH, 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['id', 'tested_positive'])
    for i, p in enumerate(test_y):
        writer.writerow([i, p])
```
